```python
from bs4 import BeautifulSoup
import numpy as np
import time
from selenium.webdriver import Edge
import pandas as pd
import random
import regex as re
```

```python
URL = 'https://www.spinny.com/used-cars-in-{}/s/'
```

```python
CITIES = [
    'delhi-ncr','bangalore'
]

Scroll_Pause = 3           # wait time after each scroll
Max_Scrolls = 30             # safety limit

car, variant, km, fuel, transmission, registration, price, location, City = [], [], [], [], [], [], [], [], []
driver = Edge()
total_start_time = time.time()
for city in CITIES:
    Target_Records = random.randint(301,380)  # minimum records to load before scraping
    driver.get(URL.format(city))
    time.sleep(5)
    city_start_time = time.time()
    print(f"\n Loading city: {city}")


    last_count = 0
    scroll_count = 0
    last_height=driver.execute_script('return document.body.scrollHeight')
    while scroll_count < Max_Scrolls:
        soup = BeautifulSoup(driver.page_source, 'lxml')
        cards = soup.find_all('div', class_="CarListingCardDetail__productDetailContainer")
        current_count = len(cards)

        print(f"   -> Loaded cards: {current_count}")

        # Stop if no new cards OR target reached
        if current_count == last_count or current_count >= Target_Records:
            break

        last_count = current_count

        last_card = driver.find_elements("css selector", "div.CarListingCardDetail__productDetailContainer")[-1]

        driver.execute_script("arguments[0].scrollIntoView({block:'center'});",last_card)

        time.sleep(Scroll_Pause)
        new_height=driver.execute_script("return document.body.scrollHeight")
        scroll_count += 1
        if last_height==new_height:
            break
        else:
            last_height=new_height
    print(f"   Finished loading {current_count} cards")
    city_records = 0
    for c in cards:
        car.append(c.find('span', class_="ListingBrandModelDetail__make").text if c.find('span', class_="ListingBrandModelDetail__make") else np.nan)
        price.append(c.find('span', class_="ListingBrandModelDetail__priceWithRupeeSymbol").text if c.find('span', class_="ListingBrandModelDetail__priceWithRupeeSymbol") else np.nan)

        variant.append(c.find('span', class_="ListingPricingDetail__variant").text if c.find('span', class_="ListingPricingDetail__variant") else np.nan)
        
        info = c.find('ul', class_="CarListingCardDetail__more")
        a=info.find_all('li')
        km.append(a[0].text if len(a) > 0 else np.nan)
        fuel.append(a[1].text if len(a) > 1 else np.nan)
        transmission.append(a[2].text if len(a) > 2 else np.nan)
        registration.append(a[3].text if len(a) > 3 else np.nan)
        #price.append(c.find('p', class_="sc-bcXHqh hvRpEM").text if c.find('p', class_="sc-bcXHqh hvRpEM") else np.nan)
        location.append(c.find('span', class_="HubDetails__hubAddress HubDetails__truncateHubDesktop").text if c.find('span', class_="HubDetails__hubAddress HubDetails__truncateHubDesktop") else np.nan)
        City.append(city)
        city_records += 1

    city_time = time.time() - city_start_time
    print(f"   Records scraped: {city_records}")
    print(f"   Time taken: {city_time:.2f} seconds")
    
driver.quit()
df1 = pd.DataFrame({
    'car': car,
    'variant': variant,
    'km': km,
    'fuel': fuel,
    'transmission': transmission,
    'registration': registration,
    'price': price,
    'location': location,
    'city' : City
})
df1.drop_duplicates(inplace=True,ignore_index=True)
df1.dropna(ignore_index=True).to_csv('Raw_Data1.csv', index=False)


total_time = time.time() - total_start_time
print(f"\nTOTAL TIME: {total_time:.2f} seconds")
print(f"TOTAL RECORDS: {len(df1)}")
print("Scraping completed successfully")

```
